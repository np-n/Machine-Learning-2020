Applying augmentation i above image

```
datagen = ImageDataGenerator(rotation_range=20,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             rescale =1./255,
                             shear_range = 0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             fill_mode='nearest')

img=load_img(sample_image_path+'shringkala2.jpeg')
img = img_to_array(img)# this is a Numpy array with shape (1042, 735, 3)
img.shape
img = img.reshape((1,)+ img.shape)# this is a Numpy array with shape (1042, 735, 3,1)
img.shape
```

```
# Generate 20 images
i = 0
for batch in datagen.flow(img,batch_size=1,save_to_dir='/content/drive/My Drive/saved_images/augmented_images/shrinkhala_augmented/',save_prefix='shrinkhala',save_format='png'):
  i+=1
  if (i == 20):
    break
```

```
# Path to store image
augmented_image_path = '/content/drive/My Drive/saved_images/augmented_images/shrinkhala_augmented/'
```

```

# Displaying generated images

import matplotlib.image as mpimg # image module for image reading
import glob
images = []
for img_path in glob.glob('//content/drive/My Drive/saved_images/augmented_images/shrinkhala_augmented/*.png'):
    images.append(mpimg.imread(img_path))

plt.figure(figsize=(15,15))
columns = 5
for i, image in enumerate(images):
    plt.subplot(len(images) / columns + 1, columns, i + 1)
    plt.imshow(image)
```
