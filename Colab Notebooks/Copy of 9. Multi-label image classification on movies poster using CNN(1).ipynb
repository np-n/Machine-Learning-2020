```
# Reading .csv file where all image address are presented as ID and description are given
data = pd.read_csv('/content/Movies-Poster_Dataset/train.csv')
data.head()
```

```
# Not all image have information in the data file
data.shape[0]
```

```
# To convert image into similar size
# let define height and width 
img_width = 350
img_height =350

# creating empty array to add image
x = []

for i in tqdm(range(data.shape[0])):
  # Getting path of all image
  path = '/content/Movies-Poster_Dataset/Images/'+ data['Id'][i] + '.jpg'
  # loading all image in img shaping by target_size
  img = image.load_img(path,target_size=(img_width,img_height))
  # To compute over images we are converting images to pixel matrix
  img = image.img_to_array(img)
  # Neural networks perform well on the normalized image
  img = img/255.0

  # Appending all normalized image (matrix ) into  multidimensional array
  x.append(img)

# Converting array to numpy array
x = np.array(x)
```
